## Nice function solution

```python
import pandas as pd
```

```python
def get_data():
    clust = pd.read_csv('E-MTAB-7365.clusters.csv')
    expr = pd.read_csv('E-MTAB-7365.csv', index_col=0)
    clust = clust.loc[clust['sel.K'] == True] # get a one row data frame for the row where sel.K is true
    k = int(clust['K']) # save K value
    columns = []
    [columns.append(i) for i in range(1,k+1)] # build a list of the columns needed to build the avgs data frame
    avgs = pd.DataFrame(columns=columns, index = expr.index) # initiate and df filled with NaNs with the needed column and index names
    return clust, expr, avgs

def avgs_df(clust, expr, avgs):
    for i in avgs.columns:
        cells = list(clust[clust.columns[clust.isin([i]).all()]].columns[1:]) # list of cells in given cluster
        avgs[i] = expr[cells].mean(axis=1) # calculate the average for every row using only columns from cells list
    return avgs
        
def main():
    clust, expr, avgs = get_data()
    averages_by_cluster=avgs_df(clust, expr, avgs)
    return averages_by_cluster
```

```python
a=main()
a.head()
```
