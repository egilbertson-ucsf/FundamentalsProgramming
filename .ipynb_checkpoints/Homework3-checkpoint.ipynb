Redo HW1, but using pandas to make life oh so much better

```python
import pandas as pd
```

```python
def get_data():
    clust = pd.read_csv('E-MTAB-7365.clusters.csv')
    expr = pd.read_csv('E-MTAB-7365.csv', index_col=0)
    clust = clust.loc[clust['sel.K'] == True]
    k = int(clust['K'])
    columns = []
    [columns.append(i) for i in range(1,k+1)]
    avgs = pd.DataFrame(columns=columns, index = expr.index)
    return clust, expr, avgs

def avgs_df(clust, expr, avgs):
    for i in avgs.columns:
        cells = list(clust[clust.columns[clust.isin([i]).all()]].columns[1:])
        avgs[i] = expr[cells].mean(axis=1)
    return avgs
        
def main():
    clust, expr, avgs = get_data()
    averages_by_cluster=avgs_df(clust, expr, avgs)
    return averages_by_cluster
```

```python
a=main()
a.head()
```
